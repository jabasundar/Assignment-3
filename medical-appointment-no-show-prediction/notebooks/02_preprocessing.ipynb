```python
df = pd.read_csv("../data/raw/Medical_appointment_data.csv")

print(df.shape)

df.head()

# Fill categorical missing values

categorical_cols = [
    "specialty",
    "disability",
    "place"
]

for col in categorical_cols:
    df[col] = df[col].fillna("Unknown")

# Fill numerical missing values

numerical_cols = [
    "age",
    "average_temp_day",
    "average_rain_day",
    "max_temp_day",
    "max_rain_day"
]

for col in numerical_cols:
    df[col] = df[col].fillna(df[col].median())

print(df.isnull().sum())

# Convert target variable

df["no_show"] = df["no_show"].map({
    "yes": 1,
    "no": 0
})

df["no_show"].value_counts()

# Age Groups

df["age_group"] = pd.cut(
    df["age"],
    bins=[0, 12, 18, 35, 60, 100],
    labels=["Child", "Teen", "Young Adult", "Adult", "Senior"]
)

# Weekend Feature

df["is_weekend_shift"] = df["appointment_shift"].apply(
    lambda x: 1 if x.lower() == "evening" else 0
)

# Health Risk Score

health_cols = [
    "under_12_years_old",
    "over_60_years_old",
    "patient_needs_companion"
]

df["health_risk_score"] = df[health_cols].sum(axis=1)

df.head()

label_encoders = {}

categorical_features = df.select_dtypes(
    include=["object", "category"]
).columns

for col in categorical_features:
    
    le = LabelEncoder()
    
    df[col] = le.fit_transform(df[col].astype(str))
    
    label_encoders[col] = le

print("Encoding Completed")

df.to_csv("../data/processed/processed_medical_data.csv", index=False)

print("Processed dataset saved successfully")
```
